```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = sns.load_dataset('titanic')
df.to_csv('titanic.csv', index=False)  # optional, local CSV bana dega
print(df.head())
print(df.tail())
print(df.info())
print(df.describe())
```

```python
print(df.isnull().sum())
df['age'].fillna(df['age'].median(), inplace=True)   # Example imputation
df.dropna(subset=['embarked'], inplace=True)         # Example removal

df.drop_duplicates(inplace=True)
sns.boxplot(x=df['age'])
plt.show()
```

```python
from scipy import stats
import matplotlib.pyplot as plt

print("Mean:", df['age'].mean())
print("Median:", df['age'].median())
print("Trimmed Mean:", stats.trim_mean(df['age'].dropna(), 0.1))
print("Range:", df['age'].max() - df['age'].min())
print("Variance:", df['age'].var())
print("Std Dev:", df['age'].std())

df['age'].hist()
plt.show()

sns.boxplot(x=df['age'])
plt.show()
```

```python
print(df['sex'].value_counts())
print("Mode:", df['sex'].mode()[0])
sns.countplot(x='sex', data=df)
```

```python
sns.scatterplot(x='age', y='fare', data=df)
plt.show()

sns.heatmap(df.select_dtypes(include=['number']).corr(), annot=True, cmap="coolwarm")
plt.show()
```

```python
sns.boxplot(x='sex', y='age', data=df)
plt.show()

sns.barplot(x='pclass', y='fare', data=df)
plt.show()
```
